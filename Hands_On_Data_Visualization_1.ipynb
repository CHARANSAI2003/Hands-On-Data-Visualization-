# Introduction to Matplotlib

There are several different ways of using Matplotlib. In this notebook, we will use the main object-oriented interface.

A single plot is a hierarchy of matplotlib objects. In order to create a plot, we need to create a Figure object and an Axes object. We can do this using `plt.subplots.`
 - The Figure object is a container that holds everything that you see on the page.
 - The Axes is the part of the page that holds the data.

A figure can hold one or more axes.



Given below is a Figure and Axes. We haven't added any data yet.


```python
plt.style.use('default') #Setting the style
```

```python
fig, ax = plt.subplots()

plt.show()
```

## Adding data to an Axes object


```python
# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

# Plot MLY-PRCP-NORMAL from seattle_weather against the MONTH
ax.plot(delhi_weather.index, delhi_weather['MLY-PRCP-NORMAL'])

# Plot MLY-PRCP-NORMAL from austin_weather against MONTH
ax.plot(kolkata_weather.index, kolkata_weather['MLY-PRCP-NORMAL'])

# Call show to display the resulting plot
plt.show()
```

## Adding Title & Labels

```python
fig, ax = plt.subplots()

# Plot Delhi data
ax.plot(delhi_weather.index, delhi_weather['MLY-PRCP-NORMAL'])

# Plot Kolkata data,
ax.plot(kolkata_weather.index, kolkata_weather['MLY-PRCP-NORMAL'])

# Add the x-axis label
ax.set_xlabel('Time(months)')

# Add the y-axis label
ax.set_ylabel('Precipitation (inches)')

# Add the title
ax.set_title('Weather Patterns in Delhi and Kolkata')

# Call show to display the resulting plot
plt.show()
```

# Dual Axis Line Charts

```python
with plt.style.context('seaborn'):

  fig, ax = plt.subplots()

  # Plot the CO2 variable in blue
  ax.plot(climate_change.index, climate_change['co2'], label='CO2 (ppm)')

  # Plot the relative temperature in red
  ax.plot(climate_change.index, climate_change['relative_temp'], label='Relative Temp (C)')

  ax.set_xlabel('Time (years)')
  ax.set_title('CO2 and Temperature over the Years')

  ax.legend()

  plt.show()
```

## Plotting two variables - twin axes

```python
with plt.style.context('seaborn'):

  fig, ax = plt.subplots()

  # Plot CO2 in blue
  ax.plot(climate_change.index, climate_change['co2'])
  ax.set_xlabel('Time (years)')
  ax.set_ylabel('CO2 Levels (ppm)')

  # Create a twin Axes that shares the x-axis
  ax2 = ax.twinx()

  # Plot the relative temperature in red
  ax2.plot(climate_change.index, climate_change['relative_temp'], c='#e41a1c')
  ax2.set_ylabel('Relative Temperature (Celsius)')

  ax.set_title('CO2 and Temperature over the Years', fontsize=19, pad=15)

  plt.show()
```

# Customizations

## Colors

```python
with plt.style.context('seaborn'):

  fig, ax = plt.subplots()

  # Plot CO2 in blue
  ax.plot(climate_change.index, climate_change['co2'], c='#377eb8')
  ax.set_xlabel('Time (years)', fontsize=17, labelpad=15)
  ax.set_ylabel('CO2 Levels (ppm)', c='#377eb8', fontsize=17, labelpad=15)
  ax.tick_params(axis='y', labelsize=15, colors='#377eb8')

  # Create a twin Axes that shares the x-axis
  ax2 = ax.twinx()

  # Plot the relative temperature in red
  ax2.plot(climate_change.index, climate_change['relative_temp'], c='#e41a1c')
  ax2.set_ylabel('Relative Temperature (Celsius)', c='#e41a1c', fontsize=17, labelpad=15)
  ax2.tick_params(axis='y', labelsize=15, colors='#e41a1c')

  ax.set_title('CO2 and Temperature over the years', fontsize=18, pad=15, c='darkgreen')

  plt.show()
```

The list of named colors in matplotlib can be viewed [here](https://matplotlib.org/stable/gallery/color/named_colors.html).

## Annotations

```python
with plt.style.context('seaborn'):

  fig, ax = plt.subplots()

  # Plot the relative temperature data
  ax.plot(climate_change.index, climate_change['relative_temp'], c='#377eb8')

  ax.set_xlabel('Time (years)', fontsize=17, labelpad=15)
  ax.set_ylabel('Relative Temperature (Celsius)', fontsize=17, labelpad=15)
  ax.tick_params(axis='both', labelsize=15)

  # Annotate the date at which temperatures exceeded 1 degree
  ax.annotate('>1 degree', fontsize=14,
              xy=(pd.to_datetime('2015-10-06'), 1),
              xytext=(pd.to_datetime('2008-10-06'), 0.1),
              arrowprops={'arrowstyle':'simple', 'color':'#e41a1c'})

  plt.show()
```

For more on annotations, refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html.


# Small Multiples

Adding too much data to a single plot can make it hard to read.
  - Instead, we can use multiple sub plots that show different parts of the data, making it easier for comparision. These multiple sub plots are called small multiples.

## Creating Subplots

- Up till now, we've called `plt.subplots` with no inputs, so it created a single subplot.

- Here, we are creating a figure object with 3 rows of subplots and 2 columns.
   - In this case, the variable `ax` isn't a single Axes object. It's an array of Axes objects with a shape of 3 by 2.

```python
fig, ax = plt.subplots(3,2)

plt.show()
```

## Adding data to Subplots

```python
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Adjust the size of the figure
fig.set_size_inches([10, 6])

# Addressing the top left Axes as index 0, 0, plot month and Delhi precipitation
ax[0, 0].plot(delhi_weather.index, delhi_weather['MLY-PRCP-NORMAL'])
ax[0, 0].set_title('Delhi Precipitation')

# In the top right (index 0,1), plot month and Delhi temperatures
ax[0, 1].plot(delhi_weather.index, delhi_weather['MLY-TAVG-NORMAL'])
ax[0, 1].set_title('Delhi Temperature')

# In the bottom left (1, 0) plot month and Kolkata precipitations
ax[1, 0].plot(kolkata_weather.index, kolkata_weather['MLY-PRCP-NORMAL'])
ax[1, 0].set_title('Kolkata Precipitation')

# In the bottom right (1, 1) plot month and Kolkata temperatures
ax[1, 1].plot(kolkata_weather.index, kolkata_weather['MLY-TAVG-NORMAL'])
ax[1, 1].set_title('Kolkata Temperature')

#Set the title for the entire subplot
fig.suptitle('Weather in Delhi and Kolkata', fontsize=15, fontweight ='bold')

# Automatically adjusts the padding between and around subplots
fig.tight_layout()

# Adjust the positioning
fig.subplots_adjust(top=0.85, wspace=0.15, hspace=0.5)

plt.show()
```

For more on `subplots_adjust`, refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html.

## Subplots with a Shared y-axis

Using `sharey=True` ensures that both subplots have the same range of y-axis values, so that comparision is easier.



```python
# Create a figure and an array of axes: 2 rows, 1 column with shared y axis
fig, ax = plt.subplots(2, 1, sharey=True)

# Plot Delhi precipitation data in the top axes
ax[0].plot(delhi_weather.index, delhi_weather['MLY-PRCP-NORMAL'], color = '#377eb8')

# Plot Kolkata precipitation data in the bottom axes
ax[1].plot(kolkata_weather.index, kolkata_weather['MLY-PRCP-NORMAL'], color = '#e41a1c')

fig.suptitle('Precipitation in Delhi and Kolkata', fontsize=15, fontweight='bold')

plt.show()
```

Note that when there is only one row or one column of plots, the resulting **axes array will be one-dimensional**, so we only have to **provide one index to access the elements**.
